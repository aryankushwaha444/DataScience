### Plotting the dataset

```python
from sklearn.datasets import make_regression
import numpy as np
input_,target_ = make_regression(n_samples=500,n_features=1,n_informative=5,n_targets=1)
input_= input_*2*(np.random.randn(1))
```

```python
import matplotlib.pyplot as plt
plt.scatter(input_,target_)
plt.xlabel('input')
plt.ylabel('target')

# Text(0,0.5,'target')
```

```python
input_= input_*2*(np.random.randint(2))
plt.scatter(input_,target_)
plt.xlabel('input')
plt.ylabel('target')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


train_input, test_input, train_target, test_target = train_test_split(input_, target_, test_size=0.2, random_state=42)

# Check the shapes of the splits
print(train_input.shape) 
print(test_input.shape)  
print(test_target.shape)  
print(train_target.shape) 


LR = LinearRegression()
LR.fit(train_input, train_target)


# y_pred = LR.predict(test_input)
test_target = LR.predict(test_input)
test_target
```

```python
from sklearn.metrics import accuracy_score
# accuracy = accuracy_score(test_target,test_input)
correct = 0
for i in range(len(test_target)):
    if test_target[i] == test_input[i]:
        correct +=1
acc = (correct/len(test_target))*100
acc        
```
