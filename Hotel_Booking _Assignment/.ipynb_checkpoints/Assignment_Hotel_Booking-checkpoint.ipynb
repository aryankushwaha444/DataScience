```python
import pandas as pd

# Load the dataset
df = pd.read_csv('hotel_bookings.csv')

# Display the first few rows of the dataset
# df.head(4) top-down 4 rowa 
# df.tail() last-up rows
df

```

### 7. Feature Engineering:
1. Create new features based on existing columns, such as calculating the duration
of stay from stays_in_weekend_nights and stays_in_week_nights.
2. Derive new features from date columns like arrival_date_year,
arrival_date_month, arrival_date_week_number, and arrival_date_day_of_month
(e.g., season, month name, day of the week).

```python
# Create new feature: Duration of stay
df_cleaned['duration_of_stay'] = df_cleaned['stays_in_weekend_nights'] + df_cleaned['stays_in_week_nights']

# Derive features from date columns
df_cleaned['arrival_month_name'] = df_cleaned['arrival_date'].dt.month_name()
df_cleaned['arrival_day_of_week'] = df_cleaned['arrival_date'].dt.day_name()

# Verify the new features
print(df_cleaned[['duration_of_stay', 'arrival_month_name', 'arrival_day_of_week']].head())

# Additional derived features
df_cleaned['arrival_week_number'] = df_cleaned['arrival_date'].dt.isocalendar().week
df_cleaned['arrival_season'] = df_cleaned['arrival_date'].dt.month % 12 // 3 + 1
df_cleaned['arrival_season'] = df_cleaned['arrival_season'].map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

# Verify additional features
print(df_cleaned[['arrival_week_number', 'arrival_season']].head())
```
