```python
years = np.arange(2012,2015)
category1_values  =[2,5,9]
category2_values = [7,6,3]

plt.figure()
plt.bar(years - 0.2 , category1_values , color='blue' , edgecolor = 'none' , width = 0.4 , align = 'center' , label = 'y1')
# plt.show()

# plt.figure()
plt.bar(years + 0.2 , category2_values , color='orange' , edgecolor = 'none' , width = 0.4 , align = 'center' , label = 'y1')
plt.xticks(years,[str(year) for year in years])
plt.legend()
plt.show()
```

```python
dist1 = np.random.normal(42, 7, 1000)
dist2 = np.random.normal(59, 3, 1000)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(dist1)
plt.title('dist1')
plt.subplot(1, 2, 2)
plt.scatter(dist2, dist1)
plt.xlabel('dist2')
plt.ylabel('dist1')
plt.tight_layout()
```
